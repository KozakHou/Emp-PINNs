```python
# Concatenate all dataframes into one
df = pd.concat([df_1996, df_1997, df_1998, df_1999, df_2000, df_2001, df_2002, df_2003,\
    df_2004, df_2005, df_2006, df_2007, df_2008, df_2009, df_2010, df_2011, df_2012, df_2013, \
        df_2014, df_2015, df_2016, df_2017], ignore_index=True)
df
```

```python
df = df.dropna()
df
```

```python
# combine year, month, day, hour, minute into one column
df['date'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute']])
df = df.drop(['year', 'month', 'day', 'hour', 'minute'], axis=1)
df
```

```python
import numpy as np
df['sw_vt'] = np.sqrt(df['vx']**2 + df['vy']**2 + df['vz']**2)
df['Dp'] = df['sw_vt']**2*df['np']*1.6726*10**(-6)
df
```

```python
df = df.drop(['np', 'tp', 'ae', 'al', 'au', 'symh'], axis=1)
df
```

```python
# calculate the vx vy's angle
df['angle'] = np.arctan2(df['vy']*df['xx'] - df['vz']*df['vy'], df['vx']*df['xx'] + df['vy']*df['yy'] + df['vz']*df['zz'])
df
```

```python
df['angle_deg'] = df['angle']*180/np.pi
df
```
