```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('outcomes.csv')
df
```

```python
df['theta_deg'] = np.degrees(df['theta'])
df
```

# for abs(theta_deg) < 45

```python
df_lt45 = df[abs(df['theta_deg']) < 45]
from sklearn.metrics import mean_squared_error
# calculate the RMSE for all r
print('RMSE of r_pred_shue: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_pred_shue'])))
print('RMSE of r_pred_hou: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_pred_hou'])))
print('RMSE of pinn_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_pinn'])))
print('RMSE of nn_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_nn'])))
print('RMSE of nn_l1_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_nn_l1'])))
print('RMSE of nn_l2_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_nn_l2'])))
print('RMSE of nn_elastic_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_nn_elastic'])))
print('RMSE of pinn_l1_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_pinn_l1'])))
print('RMSE of pinn_l2_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_pinn_l2'])))
print('RMSE of pinn_elastic_pred: ', np.sqrt(mean_squared_error(df_lt45['r'], df_lt45['r_pinn_elastic'])))
```

# for 120 > abs(theta_deg) > 45

```python
df_45_120 = df[(abs(df['theta_deg']) >= 45) & (abs(df['theta_deg']) < 120)]
# calculate the RMSE for all r
print('RMSE of r_pred_shue: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_pred_shue'])))
print('RMSE of r_pred_hou: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_pred_hou'])))
print('RMSE of pinn_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_pinn'])))
print('RMSE of nn_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_nn'])))
print('RMSE of nn_l1_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_nn_l1'])))
print('RMSE of nn_l2_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_nn_l2'])))
print('RMSE of nn_elastic_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_nn_elastic'])))
print('RMSE of pinn_l1_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_pinn_l1'])))
print('RMSE of pinn_l2_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_pinn_l2'])))
print('RMSE of pinn_elastic_pred: ', np.sqrt(mean_squared_error(df_45_120['r'], df_45_120['r_pinn_elastic'])))
```
