```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import emcee
plt.style.use('seaborn')
```

```python
p_df = pd.read_csv('p_df.csv')
b_df = pd.read_csv('b_df.csv')
p_df['Bz'] = -0.7338975205937526
b_df['Dp'] = 1.7335727299047798
df = pd.concat([p_df, b_df], ignore_index=True)
df
```

```python
# Define the alpha
def func_a(x, a1, a2, a3):
    return (a1 + a2 * x[0]) * (x[1] ** (1/a3))

# Define the log-likelihood function
def ln_likelihood(theta, x, y, yerr):
    model = func_a(x, theta[0], theta[1], theta[2])
    chi2 = np.sum((y - model)**2 / yerr**2)
    return -0.5 * chi2


# Define the log-prior function
def ln_prior(theta):
    # Set uniform priors for all parameters
    if -1000.0 < theta[0] < 1000.0 and -10000.0 < theta[1] < 10000.0 and -10000.0 < theta[2] < 10000.0:
        return 0.0
    else:
        return -np.inf
    
    
# Define the log-posterior function
def ln_posterior(theta, x, y, yerr):
    lp = ln_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + ln_likelihood(theta, x, y, yerr)
```

```python
# Load df
x = np.array([df['Bz'], df['Dp']])
y = np.array(df['alpha'])
yerr = np.std(y) * np.ones_like(y)

# Set up the MCMC sampler
nwalkers = 40
ndim = 3
pos = np.random.randn(nwalkers, ndim)
sampler = emcee.EnsembleSampler(nwalkers, ndim, ln_posterior, args=(x, y, yerr))

# Burn-in phase
pos, _, _ = sampler.run_mcmc(pos, 10000)

# Production phase
sampler.reset()
sampler.run_mcmc(pos, 10000)

# Get the chain and flatten it
chain = sampler.get_chain(discard=1000, flat=True)

# Get the posterior samples and uncertainties
samples_a = np.percentile(chain, [68, 70, 73], axis=0)
samples_a
```

```python
# Plot the posterior samples
fig, axes = plt.subplots(3, figsize=(10, 7), sharex=True)
labels = ["a1", "a2", "a3"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(chain[:, i], "k", alpha=0.3)
    ax.set_xlim(0, len(chain))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)
    
    
axes[-1].set_xlabel("step number")
```

```python
# Define the alpha

def func_r(x, a4, a5, a6, a7, a8, a9, a10, a11):
    return (a4 + a5 * np.tanh(a6 * (x[0] + a7)) + a8 * np.tanh(a9 * (x[0] + a10))) * (x[1] ** (-1/a11))

# Define the log-likelihood function
def ln_likelihood(theta, x, y, yerr):
    model = func_r(x, theta[0], theta[1], theta[2], theta[3], theta[4], theta[5], theta[6], theta[7])
    chi2 = np.sum((y - model)**2 / yerr**2)
    return -0.5 * chi2


# Define the log-prior function
def ln_prior(theta):
    if (-10 < theta[0] < 10 and -10 < theta[1] < 10 and -10 < theta[2] < 10 \
        and -20 < theta[3] < 20 and -50 < theta[4] < 50 and -50 < theta[5] < 50 \
        and -20 < theta[6] < 20 and -50 < theta[7] < 50):
        return 0.0
    else:
        return -np.inf
    
    
# Define the log-posterior function
def ln_posterior(theta, x, y, yerr):
    lp = ln_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + ln_likelihood(theta, x, y, yerr)
```

```python
# Load df
x = np.array([df['Bz'], df['Dp']])
y = np.array(df['r0'])
yerr = np.std(y) * np.ones_like(y)

# Set up the MCMC sampler
nwalkers = 75
ndim = 8
pos = np.random.randn(nwalkers, ndim)
sampler = emcee.EnsembleSampler(nwalkers, ndim, ln_posterior, args=(x, y, yerr))

# Burn-in phase
pos, _, _ = sampler.run_mcmc(pos, 1000)

# Production phase
sampler.reset()
sampler.run_mcmc(pos, 10000)

# Get the chain and flatten it
chain = sampler.get_chain(discard=1000, flat=True)

# Get the posterior samples and uncertainties
samples_r = np.percentile(chain, [68, 70, 73], axis=0)
samples_r
```

```python
# Plot the posterior samples
fig, axes = plt.subplots(8, figsize=(10, 7), sharex=True)
labels = ["a0", "a1", "a2", "a3", "a4", "a5", "a6", "a7"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(chain[:, i], "k", alpha=0.3)
    ax.set_xlim(0, len(chain))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)
    
    
axes[-1].set_xlabel("step number");
```

```python
sample_r_0 = samples_r[0]
sample_r_50 = samples_r[1]
sample_r_100 = samples_r[2]


sample_a_0 = samples_a[0]
sample_a_50 = samples_a[1]
sample_a_100 = samples_a[2]
```

```python
def r0_0(Bz, P):
    return (sample_r_0[0] + sample_r_0[1] * np.tanh(sample_r_0[2] * (Bz + sample_r_0[3])) + sample_r_0[4] * np.tanh(sample_r_0[5] * (Bz + sample_r_0[6]))) * (P ** (-1/sample_r_0[7]))

def alpha_0(Bz, P):
    return (sample_a_0[0] + sample_a_0[1] * Bz) * (P ** (1/sample_a_0[2]))
```

```python
def r0_50(Bz, P):
    return (sample_r_50[0] + sample_r_50[1] * np.tanh(sample_r_50[2] * (Bz + sample_r_50[3])) + sample_r_50[4] * np.tanh(sample_r_50[5] * (Bz + sample_r_50[6]))) * (P ** (-1/sample_r_50[7]))
# define alpha
def alpha_50(Bz, P):
    return (sample_a_50[0] + sample_a_50[1] * Bz) * (P ** (1/sample_a_50[2]))
```

```python
# define r0
def r0_100(Bz, P):
    return (sample_r_100[0] + sample_r_100[1] * np.tanh(sample_r_100[2] * (Bz + sample_r_100[3])) + sample_r_100[4] * np.tanh(sample_r_100[5] * (Bz + sample_r_100[6]))) * (P ** (-1/sample_r_100[7]))

def alpha_100(Bz, P):
    return (sample_a_100[0] + sample_a_100[1] * Bz) * (P ** (1/sample_a_100[2]))
```

```python
def r_0(Bz, P, theta):
    return r0_0(Bz, P) * ((2/(1+np.cos(theta)))**(alpha_0(Bz, P)))

def r_50(Bz, P, theta):
    return r0_50(Bz, P) * ((2/(1+np.cos(theta)))**(alpha_50(Bz, P)))

def r_100(Bz, P, theta):
    return r0_100(Bz, P) * ((2/(1+np.cos(theta)))**(alpha_100(Bz, P)))
```

```python
# make a 2D plane of r_pred_hou for different Bz and P
Bz = np.linspace(-20, 20, 100)
P = np.linspace(0.5, 60, 100)

Bz, P = np.meshgrid(Bz, P)

r_pred_mcmc_0 = r_0(Bz, P, 0)
r_pred_mcmc_50 = r_50(Bz, P, 0)
r_pred_mcmc_100 = r_100(Bz, P, 0)
```

```python
# load mp.csv
mp = pd.read_csv('mp.csv')
mp = mp.drop(columns = ['year1', 'month1', 'day1', 'hour1','minute1', 'second1', \
    'year2', 'month2', 'day2', 'hour2', 'minute2', 'second2', 'probe', 'inbound'])
mp.dropna(inplace = True)
mp.reset_index(drop = True, inplace = True)
# Define a rotation matrix to rotate 4 degree to mp_xx, mp_yy, mp_zz
def rotate_4(x, y, z):
    theta = np.pi/45
    R = np.array([[np.cos(theta), -np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0, 1]])
    return np.dot(R, np.array([x, y, z]))

# Rotate 4 degree to mp_xx, mp_yy, mp_zz
mp_xx, mp_yy, mp_zz = rotate_4(mp['mp_xx'], mp['mp_yy'],mp['mp_zz'])

# Concatenate the rotated mp_xx, mp_yy, mp_zz to the original data
rr = pd.DataFrame({'mp_xx_r': mp_xx, 'mp_yy_r': mp_yy, 'mp_zz_r': mp_zz})
mp = pd.concat([mp, rr], axis = 1)
mp.drop(columns = ['mp_xx', 'mp_yy', 'mp_zz', 'she_bx', 'she_by', 'she_bz', 'sph_bx', 'sph_by', 'sph_bz', 'imf_bx', 'imf_by'], inplace = True)
mp['r_real'] = np.sqrt(mp['mp_xx_r']**2 + mp['mp_yy_r']**2 + mp['mp_zz_r']**2)
mp['P'] = mp['sw_vt']**2*mp['sw_np']*1.67*10**(-6)
mp['theta'] = np.arctan2(mp['mp_yy_r'], mp['mp_xx_r'])
mp['r_MCMC_40'] = r0_0(mp['imf_bz'], mp['P']) * ((2/(1+np.cos(mp['theta'])))**(alpha_0(mp['imf_bz'], mp['P'])))
mp['r_MCMC_50'] = r0_50(mp['imf_bz'], mp['P']) * ((2/(1+np.cos(mp['theta'])))**(alpha_50(mp['imf_bz'], mp['P'])))
mp['r_MCMC_60'] = r0_100(mp['imf_bz'], mp['P']) * ((2/(1+np.cos(mp['theta'])))**(alpha_100(mp['imf_bz'], mp['P'])))
# RMSE for MCMC against r_real
print('RMSE for MCMC against r_real', np.sqrt(np.mean((mp['r_MCMC_40'] - mp['r_real'])**2)))
print('RMSE for MCMC against r_real', np.sqrt(np.mean((mp['r_MCMC_50'] - mp['r_real'])**2)))
print('RMSE for MCMC against r_real', np.sqrt(np.mean((mp['r_MCMC_60'] - mp['r_real'])**2)))
```
